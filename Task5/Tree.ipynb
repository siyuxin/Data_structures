```python
# 
# Nodeelem
# lchildrchild
class Node(object):
    
    def __init__(self,elem = -1,lchild = None,rchild = None):
        self.elem = elem
        self.lchild = lchild
        self.rchild = rchild
        
```

```python
# ,
# root
class Tree(object):
    
    def __init__(self,root = None):
        self.root = root
        
    def add(self,elem):
        """"""
        node = Node(elem)
         #
        if self.root == None:
            self.root = node
        else:
            queue = []
            queue.append(self.root)
            #
            while queue:
                #
                cur = queue.pop(0)
                if cur.lchild == None:
                    cur.lchild = node
                    return 
                elif cur.rchild == None:
                    cur.rchild = node
                    return 
                else:
                    #
                    queue.append(cur.lchild)
                    queue.append(cur.rchild)
    def preorder(self,root):
            if root == None:
                return 
            print(root.elem)
            self.preorder(root.lchild)
            self.preorder(root.rchild)
```
